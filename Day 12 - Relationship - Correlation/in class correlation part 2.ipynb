```python
df2 = sns.load_dataset('titanic')

df2.head(10)
```

```python
df2[['fare','survived','age','pclass','sibsp','parch']].cov().round(2)

co =['fare','survived','age','pclass','sibsp','parch']
covariance = df2[co].cov()
covariance.round(2)
```
